```python

from transformers import pipeline

classifier = pipeline("zero-shot-classification",
                      model="facebook/bart-large-mnli")

sequence_to_classify = "This is a course about the Transformers library"
candidate_labels = ["education", "politics", "business"]

result = classifier(sequence_to_classify, candidate_labels)
result
```

```python


classifier = pipeline("ner", grouped_entities=True)

sequence = "Hugging Face Inc. is a company based in New York City. Its headquarters are in DUMBO, therefore very close to the Manhattan Bridge."

result = classifier(sequence)
result
```

```python
from transformers import AutoModel, AutoModelForSequenceClassification

# 'model' has less number of parameters as compared to 'cmodel'
# 'cmodel' doesn't have to be finetuned.
# only 'cmodel' can be used directly for predicting if a sentence has positive sentiment or negative sentiment (after training).

model_checkpoint = "distilbert-base-uncased-finetuned-sst-2-english"
model = AutoModel.from_pretrained(model_checkpoint)
cmodel = AutoModelForSequenceClassification.from_pretrained(model_checkpoint)

print(f"Number of parameters in 'model': {model.num_parameters()}")
print(f"Number of parameters in 'cmodel': {cmodel.num_parameters()}")
```
